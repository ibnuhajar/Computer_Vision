```
import cv2 as cv
import matplotlib.pyplot as plt

citra = cv.imread('peppers.png')
histo1 = cv.calcHist([citra], [0], None, [256], [0,256])
histo11 = cv.calcHist([citra], [1], None, [256], [0,256])
histo12 = cv.calcHist([citra], [2], None, [256], [0,256])

plt.plot(histo1)
plt.plot(histo11)
plt.plot(histo12)
plt.show()
```
