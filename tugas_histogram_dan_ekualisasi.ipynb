```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

citra = cv.imread('gembala.png')
histo1 = cv.calcHist([citra], [0], None, [256], [0,256])

plt.plot(histo1)
plt.show()
```

```
citra = cv.imread('gembala.png', 0)
ekual = cv.equalizeHist(citra)
hasil = np.hstack((citra, ekual))

# Tampilkan hasilnya
cv2_imshow( hasil)

cv.waitKey(0)
```
