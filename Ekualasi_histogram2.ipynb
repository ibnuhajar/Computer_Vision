```
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

citra = cv.imread('gembala.png',0)
clahe = cv.createCLAHE(clipLimit = 12,
                       tileGridSize = (16,16))

ekual = clahe.apply(citra)
hasil = np.hstack((citra, ekual))

# Tampilkan hasilnya
cv2_imshow(hasil)

cv.waitKey(0)
```
