```
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

citra = cv.imread('taipei101.png')
lab = cv.cvtColor(citra, cv.COLOR_BGR2LAB)

kanalLAB = cv.split(lab)

clahe = cv.createCLAHE(clipLimit = 2.0,
                       tileGridSize = (8,8))

kanalLAB[0] = clahe.apply(kanalLAB[0])

lab = cv.merge(kanalLAB)

bgr = cv.cvtColor(lab,cv.COLOR_LAB2BGR)

hasil = np.hstack((citra,bgr))

# Tampilkan hasilnya
cv2_imshow(hasil)

cv.waitKey(0)
```
