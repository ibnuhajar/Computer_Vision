```
citra = cv.imread('ibnuhajar.jpg')
copy = citra.copy()

tebal = 20
warna = [125,110,140]

jlh_baris = copy.shape[0]
jlh_kolom = copy.shape[1]

# bingkai atas
for baris in range(tebal):
  for kolom in range(jlh_kolom):
    copy[baris,kolom] = warna

# bingkai bawah
for baris in range(jlh_baris - tebal - 1, jlh_baris ):
  for kolom in range(jlh_kolom):
    copy[baris,kolom] = warna

# bingkai kiri
for baris in range(tebal , jlh_baris - tebal - 1 ):
  for kolom in range(tebal):
    copy[baris,kolom] = warna

# bingkai kanan
for baris in range(tebal , jlh_baris - tebal - 1 ):
  for kolom in range(jlh_kolom - tebal - 1 , jlh_kolom):
    copy[baris,kolom] = warna

cv2_imshow(citra)
print('\n')
print('\n')
cv2_imshow(copy)
```

```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow 

citra = cv.imread('ibnuhajar.jpg')
jlh_baris, jlh_kolom = citra.shape[:2]

matrix = cv.getRotationMatrix2D( (0,0) ,-15,1 )
hasil = cv.warpAffine(citra,matrix,(jlh_kolom,jlh_baris))
cv2_imshow(hasil)

# Karena jika menggunakan titik pusat (0,0) saat melakukan rotasi 
# maka rotasi gambar akan dimulai dari titik (0,0) pada kurva x,y, 
# sehingga membuat gambar tidak ter rotasi secara maksimal 
# karena dapat dikatakan gambar otomatis 
# terotasi dari sudut kiri bawah yaitu titik (0,0) pada kurva x,y.
```
