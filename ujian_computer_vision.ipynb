```
import numpy as np
import cv2 as cv
import matplotlib as ml
from google.colab.patches import cv2_imshow 

citra = cv.imread('ibnuhajar.jpg', cv.IMREAD_GRAYSCALE)             # membaca file gambar dan mengubah warna nya menjadi hitam putih
jumlah_pixel = citra.shape                                          # untuk mengetahui ukuran pixel gambar
baris = citra.shape[0]                                              # untuk mengetahui baris gambar
kolom = citra.shape[1]                                              # untuk mengetahui kolom gamabr

hasil = [jumlah_pixel, baris, kolom ]                               # di simpan kedalam variabel hasil dalam bentuk data array
for i in hasil:                                                     # kemudian di lakukan perulangan variabel hasil 
  print('Jumlah nya adalah = ',i)                                   # menampilkan output perulangan variabel hasil
print('\n')                                                         # secara berurut hasil output nya adalah 
                                                                    # 1.jumlah pixel
                                                                    # 2.jumlah baris
                                                                    # 3.jumlah kolom

print('hasil dari imread_grayscale : ')
cv2_imshow(citra)                                                   # menampilkan isi dari variabel hasil 
                                                                    # isi dari variabel hasil ialah gambar hitam putih

```
